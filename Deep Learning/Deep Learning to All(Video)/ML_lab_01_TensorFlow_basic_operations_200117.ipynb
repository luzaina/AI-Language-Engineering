## Start to TensorFlow

```
import tensorflow as tf
```

```
# constant 함수를 사용해서 constant("Hello, TensorFlow!라는 노드를 만들어 준 것)를 만들고 hello 이름에 넣어줌. 
hello = tf.constant("Hello, TensorFlow!")
```

```
# tf session 시작
sess = tf.Session()
```

```
# Hello라는 노드를 실행 # b라는 문자열은 bytes literals이다. 
print(sess.run(hello))
```

## Computational Graph

```
node1 = tf.constant(3.0, tf.float32) # node3 = node1+node2
node2 = tf.constant(4.0)
node3 = tf.add(node1, node2)
```

```
print("node1:",node1, "node2:", node2) # 결과값이 나오지 않는다.
print("node3:",node3)
```

```
sess = tf.Session()
print("sess.run(node1,node2):", sess.run([node1,node2])) #실행시키고 싶은 노드를 넣어 세션 실행
print("sess.run(node3):",sess.run(node3))
```
